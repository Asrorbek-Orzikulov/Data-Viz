```python
reviews.loc[reviews.city == "Jamaica Plain", "longitude"] = -71.1151431
reviews.loc[reviews.city == "Jamaica Plain", "latitude"] = 42.3097365
reviews[reviews.city == "Jamaica Plain"][["longitude", "latitude"]]
```

```python
reviews["category_new"] = np.nan
```

```python
gym_rows = reviews["categories"].str.contains("gym", case=False, na=False)
restaurant_rows = reviews["categories"].str.contains("restaurant", case=False, na=False)
print(np.sum(gym_rows), np.sum(restaurant_rows))
```

```python
reviews.loc[restaurant_rows, "category_new"] = "Restaurants"
reviews.loc[gym_rows, "category_new"] = "Gyms"
reviews.loc[~(restaurant_rows | gym_rows), "category_new"] = "Others"
reviews.category_new.value_counts()
```

```python
reviews.to_csv("with_json_data.csv", index=False)
```
